# Now for real cleaning

## 1: Recency Date

```python
df.RECENCY_DATE = pd.to_datetime(df.RECENCY_DATE.str.slice(stop=4) + '-' + df.RECENCY_DATE.str.slice(start=4))#.dt.strftime('%Y-%m')
type(df.RECENCY_DATE[0])
```

```python
df.head()
```

## 2: Location Type

```python
loc_types = {'M':'MFDU', 'N':'Nursing Home', 'R':'Retirement Home', 'S':'SFDU', 'T':'Trailer', 'U':'Undefined'}
df.LOCATION_TYPE = df.LOCATION_TYPE.replace(loc_types)
df.LOCATION_TYPE.head()
```

## 3: TRADELINE_COUNT

```python
df.TRADELINE_COUNT = df.TRADELINE_COUNT.astype(int)
df.TRADELINE_COUNT.head(2)
```

## Age Codes

```python
age_map = {'A' : '<25', 'B':'25-29', 'C':'30-34', 'D':'35-39', 'E':'40-44', 'F':'45-49', 'G':'50-54', 'H':'55-59', 'I':'60-64', 'J':'65+',
'K':'65-69', 'L':'70-74', 'M':'75+'}
df.HEAD_HH_AGE_CODE = df.HEAD_HH_AGE_CODE.replace(age_map)
df.HEAD_HH_AGE_CODE.head()
```
