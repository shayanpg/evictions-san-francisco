```python
import pandas as pd
import numpy as np
```

# First: Clean Geospatial Data

```python
maps = pd.read_csv('raw_data/San_Francisco_Bay_Region_2010_Census_Block_Groups.csv', dtype={'full_fips':str})
maps.head()
```

```python
maps = maps[maps.fipco == 75]
maps = maps.copy()
maps.head()
```

```python
maps.reset_index(inplace=True)
maps.head(1)
```

```python
maps = maps[['objectid', 'full_fips', 'Shape__Area', 'Shape__Length']].copy()
maps.head()
```

```python
maps.to_csv('clean_data/SF_block_group_shapes.csv')
```

# Second: Clean ACS 2018

```python
housing18 = pd.read_csv('raw_data/2018_ACS_Housing_B09019.csv').iloc[:, [0, 2]]
housing18.head()
```

```python
housing18.columns = ['geoid', 'total']
housing18['geoid'] = housing18.geoid.str.slice(start=9)
housing18.head()
```

```python
housing18.to_csv('clean_data/2018_acs_housing.csv')
```
