```python
df
```

```python
fips = '06075'
county_series = df.GE_CENSUS_STATE_2010 + df.GE_CENSUS_COUNTY
df = df[county_series == fips]
df = client.persist(df)

# df.head()
df.dask

# df = df.repartition(npartitions=df.npartitions // 100)
```

```python
df = df.repartition(npartitions=df.npartitions // 100)
df
```
