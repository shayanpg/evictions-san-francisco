# Regression

```python
df = df[df['Total Evictions'] < 175]
```

```python
def r(x, y):
    x_su = (x - np.mean(x)) / np.std(x)
    y_su = (y - np.mean(y)) / np.std(y)
    return np.mean(x_su * y_su)
print(r(df['Total Evictions'], df['children']))
print(r(df['Total Evictions'], df['married']))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df#.sum().to_dict()
df.loc[:, 'Non Payment':'Total Evictions'].mul(ev['hh_1000'], axis=0).sum() / sum(ev.hh_1000)
```

## Children Regression
1. Train-Test Split
2. Find Equation
3. Predict
4. Accuracy/MSE

```python
y_kid = np.array(df.children)
x_kid = np.array(df.loc[:, 'Non Payment':'Good Samaritan Ends'])
model = LinearRegression().fit(x_kid, y_kid)
r_sq = model.score(x_kid, y_kid)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:')
dict(zip(df.columns[1:len(df.columns) - 3], model.coef_))
```

Trying MLR again with just most powerful coefficients

```python
y_kid = np.array(df.children)
x_kid = np.array(df.loc[:, ['Owner Move In', 'Nuisance', 'Breach', 'Non Payment', 'Ellis Act WithDrawal']])
model = LinearRegression().fit(x_kid, y_kid)
r_sq = model.score(x_kid, y_kid)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:')
dict(zip(df.columns[1:len(df.columns) - 3], model.coef_))
```

SLR on total

```python
y_kid = np.array(df.children)
x_kid = np.array(df['Total Evictions']).reshape(-1, 1)
model = LinearRegression().fit(x_kid, y_kid)
r_sq = model.score(x_kid, y_kid)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
# dict(zip(df.columns[1:len(df.columns) - 3], model.coef_))
```

Polynomial on Total

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
y_kid = np.array(df.children)
# x_kid = np.array(df['Total Evictions']).reshape(-1, 1)
x_kid = np.array(df.loc[:, 'Non Payment':'Good Samaritan Ends'])
x_ = PolynomialFeatures(degree=2, include_bias=True).fit_transform(x_kid)
```

```python
model = LinearRegression(fit_intercept=False).fit(x_, y_kid)
r_sq = model.score(x_, y_kid)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

RMSE on itself

```python
y_pred = model.predict(x_)
np.sqrt(np.mean((y_pred - y_kid)**2))
```

#### Test Train Split

Quadratic Polynomial Regression on Eviction Causes predicting Percentage of Children

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.2)
len(train), len(test), len(df)
```

```python
y_train = np.array(train.children)
x_train = np.array(train.loc[:, 'Non Payment':'Good Samaritan Ends'])
x_tr = PolynomialFeatures(degree=2, include_bias=True).fit_transform(x_train)

model = LinearRegression(fit_intercept=False).fit(x_tr, y_train)
r_sq = model.score(x_tr, y_train)
print('coefficient of determination:', r_sq)
# print('intercept:', model.intercept_)
# print('slope:', model.coef_)

y_test = np.array(test.children)
x_test = np.array(test.loc[:, 'Non Payment':'Good Samaritan Ends'])
x_te = PolynomialFeatures(degree=2, include_bias=True).fit_transform(x_test)

y_pred = model.predict(x_te)
print('RMSE:', np.sqrt(np.mean((y_pred - y_test)**2)))
```
