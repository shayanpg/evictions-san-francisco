# Fix Variables

```python
# RECENCY DATE to datetime
df.RECENCY_DATE = pd.to_datetime(df.RECENCY_DATE.str.slice(stop=4) + '-' + df.RECENCY_DATE.str.slice(start=4))#.dt.strftime('%Y-%m')
# Length of Residence
df.LENGTH_OF_RESIDENCE = df.LENGTH_OF_RESIDENCE.astype(int)
# Location Type decoded
loc_types = {'M':'MFDU', 'N':'Nursing Home', 'R':'Retirement Home', 'S':'SFDU', 'T':'Trailer', 'U':'Undefined'}
df.LOCATION_TYPE = df.LOCATION_TYPE.replace(loc_types)
# Tradelines to int
df.TRADELINE_COUNT = df.TRADELINE_COUNT.astype(int)
# Children indicator to int
df.CHILDREN_IND = df.CHILDREN_IND.astype(int)
# Decode ages
age_map = {'A' : '<25', 'B':'25-29', 'C':'30-34', 'D':'35-39', 'E':'40-44', 'F':'45-49', 'G':'50-54', 'H':'55-59', 'I':'60-64', 'J':'65+',
'K':'65-69', 'L':'70-74', 'M':'75+'}
df.HEAD_HH_AGE_CODE = df.HEAD_HH_AGE_CODE.replace(age_map)
# Renter Status (1 if renter, else 0)
df.OWNER_RENTER_STATUS = (df.OWNER_RENTER_STATUS.astype(int) <= 4).astype(int)
# Home Value
df.ESTMTD_HOME_VAL_DIV_1000 = df.ESTMTD_HOME_VAL_DIV_1000.astype(int) * 1000
# Marital Status (1 if married)
df.MARITAL_STATUS = (df.MARITAL_STATUS.astype(int) > 4).astype(int)
# Vacant
df.VACANT = df.VACANT.astype(int)
# Coordinates
df.GE_LATITUDE_2010 = df.GE_LATITUDE_2010.astype(float)
df.GE_LONGITUDE_2010 = df.GE_LONGITUDE_2010.astype(float)
# Street
df.STREET_NAME = df.STREET_PRE_DIR.fillna('') + ' ' +  df.STREET_NAME + ' ' + df.STREET_SUFFIX + ' ' + df.STREET_POST_DIR.fillna('')
df.drop(labels=['STREET_PRE_DIR', 'STREET_POST_DIR', 'STREET_SUFFIX'], inplace=True, axis=1)
```

Ethnic Coding

```python
def ethnic_coding(df):
    # fix ethnic coding
    ethnicity = pd.read_csv('../../research/data/infogroup_raw/ethnicity_coding.csv')[['INFOGROUP_CODE', 'RACE']].groupby('INFOGROUP_CODE').agg(max).RACE.to_dict()
    df['Ethnicity_Code_1'] = df['Ethnicity_Code_1'].replace(ethnicity)
    return df
df = ethnic_coding(df)
df.head()
```
